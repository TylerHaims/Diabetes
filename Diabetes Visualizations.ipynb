```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2021.csv")
df_2015 = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2021.csv")
#df = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2021.csv")
#df_2015 = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
correlation_matrix = df.corr()

print(correlation_matrix)
# use the entire data. optimize for recall than accuracy. don't want false negatives, false positives ok.
```

```python
import seaborn as sns
plt.figsize=(100, 100)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# phy health, general, income, education, bmi, age?
plt.title('Unbalanced Dataset Correlation Heatmap')
plt.show()
```

```python
columns = ['Diabetes_binary','Income', 'Education', 'GenHlth', 'PhysHlth', 'BMI', 'Age', 'HighBP','HighChol','DiffWalk','HeartDiseaseorAttack','PhysActivity']
adjusted_corr_matrix = df[columns].corr()
sns.heatmap(adjusted_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# phy health, general, income, education, bmi, age?
plt.title('Unbalanced Dataset  Correlation Heatmap')
plt.show()
```

```python
columns = ['Diabetes_binary','Income', 'Education', 'GenHlth', 'PhysHlth', 'BMI', 'Age', 'HighBP','HighChol','DiffWalk','HeartDiseaseorAttack','PhysActivity']
adjusted_corr_matrix_2015 = df_2015[columns].corr()
sns.heatmap(adjusted_corr_matrix_2015, annot=True, cmap='coolwarm', fmt=".2f")
# phy health, general, income, education, bmi, age?
plt.title('Balanced Dataset Correlation Heatmap')
plt.show()
```
