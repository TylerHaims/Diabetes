```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_diab = pd.read_csv("/Users/tylerhaims/Documents/GitHub/Diabetes/Diabetes2021/diabetes_binary_health_indicators_BRFSS2021.csv")
print(len(df_diab['BMI']))




fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Scatter plot 1: BMI vs. Diabetes_binary
sns.scatterplot(data=df_diab, x='BMI', y='Diabetes_binary', ax=axes[0, 0])
axes[0, 0].set_title('BMI vs. Diabetes_binary')

# Scatter plot 2: GenHlth vs. Diabetes_binary
sns.scatterplot(data=df_diab, x='GenHlth', y='Diabetes_binary', ax=axes[0, 1])
axes[0, 1].set_title('GenHlth vs. Diabetes_binary')

# Scatter plot 3: PhysHlth vs. Diabetes_binary
sns.scatterplot(data=df_diab, x='PhysHlth', y='Diabetes_binary', ax=axes[1, 0])
axes[1, 0].set_title('PhysHlth vs. Diabetes_binary')

# Scatter plot 4: Education vs. Diabetes_binary
sns.scatterplot(data=df_diab, x='Education', y='Diabetes_binary', ax=axes[1, 1])
axes[1, 1].set_title('Education vs. Diabetes_binary')

# Scatter plot 5: Income vs. Diabetes_binary
sns.scatterplot(data=df_diab, x='Income', y='Diabetes_binary', ax=axes[2, 0])
axes[2, 0].set_title('Income vs. Diabetes_binary')

# Hide the sixth subplot
axes[2, 1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()


```
